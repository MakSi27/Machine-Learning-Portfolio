## OOP and Inheritance  Fundamentals

#### Key Points


- Derived classes inherit attributes and behaviors from parent base classes

- Overriding inherited methods allows modifying functionality in child classes

- Multiple inheritance allows a derived class to inherit from multiple parent classes

- Name mangling with double underscores enables a form of private class variables

## Reflection Questions
1. How does inheritance promote code reuse in OOP?

Inhritance promotes code reuse in OOP by letting you define common  attributes and behaviour once in a parent class and then atomatically shareing them with all child classes, so you don't need to rewrite the same code again. This makes your program cleaner , easier to maintain , and faster to build , because each child class can simply reuse the parent's logic and only add or modify what's unique to it.


2. What is an example where you would want to override an inherited method?

You’d want to override an inherited method when the child class needs the same method name but a more specific or different behavior. For example, if a parent class `Animal` has a method `sound()` that gives a general sound, a child class like `Dog` can override `sound()` to make it bark instead. This lets you keep the structure from the parent while customizing the action to match the child class’s real behavior.

3. How might you utilize inheritance in a program you build?

You might use inheritance in a program when several parts share common features but still need their own unique behavior. For example, if you build a payment system, you could create a `User` class with shared details like name and email, then create `Customer`, `Admin`, and `Seller` classes that inherit from `User`. Each child class gets the common code automatically and only adds the features that make it different. This keeps your project clean, organized, and easier to expand as it grows.

#### Inheritance code snippet.

```python
class UFC:
    def weight_class(self , weight):
        classes = {130:'lightweight',
                   165:'welterweight'}
        return classes[weight]
```

```python
class competitor(UFC):
    pass
```

```python
conor = competitor()
conor.name = "Conor McGregor"
conor.age = 28 
conor.weight = 130

nate = competitor()
nate.name = 'Nate Diaz'
nate.age = 30
nate.weight = 165
```

```python
conor.weight_class
```

```python
print(conor.weight_class(conor.weight))
```

#### Multiple Inheritance code snippet.

```python
class MMA:
    def org(self , org_name):
        orgs = {"UFC": "Ultimate fighting championship",
                "Bellator":"MMA promotion in Santa Monic, California."}
        return orgs[org_name]
```

```python
class competitorAll(UFC , MMA):
    pass
```

```python
gsp = competitorAll()
gsp.name = 'GSP'
gsp.age = 27
gsp.weight = 165
print(f"{gsp.name} is a G.O.A.T in the {gsp.weight_class(gsp.weight)} division of the {gsp.org("UFC")} ")
```
